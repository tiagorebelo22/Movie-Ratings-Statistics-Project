```python
import pandas as pd
import numpy as np
from ast import literal_eval
import scipy.stats as st
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# import data

movies = pd.read_csv("Movies.csv", converters={"genres": literal_eval,
                                              "country":literal_eval})

movies.head()
```

```python
movies.describe()

# it seems we have some missing data, as there are movies with 0 budget and 0 revenue
```

```python
movies[(movies["budget"]==0) | (movies["revenue"]==0)]
```

```python
# Correcting the budget for "Dunkirk" and the revenue for "Skyfall", using data extracted manually from IMDB

movies.loc[movies["title"]=="Dunkirk","budget"] = 100000000
movies.loc[movies["title"]=="Skyfall","revenue"] = 1142471295

movies[(movies["title"]=="Dunkirk") | (movies["title"]=="Skyfall")]
```

```python
movies.describe()

# all seems to be ok now
```

```python
"""
Does genre influence ratings?

Method: ANOVA of the average ratings in each genre

H0: Genre1_Ratings_Mean = Genre2_Ratings_Mean = Genre3_Ratings_Mean = Genre4_Ratings_Mean...
H1: Genre1_Ratings_Mean != Genre2_Ratings_Mean != Genre3_Ratings_Mean != Genre4_Ratings_Mean...
"""
```

```python
alpha = 0.05

movies_exploded_genres = movies.explode("genres")
unique_genres = list(set(movies_exploded_genres["genres"]))

genres_ratings = []

for genre in unique_genres:
    vote_average_genre = movies_exploded_genres[movies_exploded_genres["genres"]==genre]["vote_average"]
    genres_ratings.append(vote_average_genre)
    
stats, p_value = st.f_oneway(*genres_ratings)

if p_value < alpha:
    print("We can reject the null hypothesis (p-value: ",p_value,")")
else:
    print("We cannot reject the null hypothesis (p-value: ",p_value,")")

movies_exploded_genres.to_csv("Movies_exploded_genres.csv", index=False)
```

```python
"""
Do movies with higher budgets have better ratings?

Method: Independent samples T-test

Sample 1: Movies with a budget above or equal to the mean
Sample 2: Movies with a budget below the mean 

H0: Ratings_Mean1 >= Ratings_Mean2
H1: Ratings_Mean1 < Ratings_Mean2
"""
```

```python
alpha = 0.05

s1 = movies[movies["budget"] >= movies["budget"].mean()]
s2 = movies[movies["budget"] < movies["budget"].mean()]

stat, p_value = st.ttest_ind(s1["vote_average"], s2["vote_average"], equal_var = False,alternative = "less")

if p_value < alpha and stat < 0: # since it is an inequality, the sign of the statistic needs to be in the direction of the alternative
    print("We can reject the null hypothesis (stat: ",stat," | p-value: ",p_value,")")
else:
    print("We cannot reject the null hypothesis (p-value: ",p_value,")")
    
print("Mean all movies:",movies["vote_average"].mean())
print("Mean sample 1:",s1["vote_average"].mean())
print("Mean sample 2:",s2["vote_average"].mean())

```

```python
"""
Do movies with higher ratings have better revenues?

Method: Independent samples T-test

Sample 1: Movies with ratings above or equal to the mean
Sample 2: Movies with ratings below the mean 

H0: Revenue_Mean1 >= Revenue_Mean2
H1: Revenue_Mean1 < Revenue_Mean2
"""
```

```python
alpha = 0.05

s1 = movies[movies["vote_average"] >= movies["vote_average"].mean()]
s2 = movies[movies["vote_average"] < movies["vote_average"].mean()]

stat, p_value = st.ttest_ind(s1["revenue"], s2["revenue"], equal_var = False,alternative = "less")

if p_value < alpha and stat < 0: # since it is an inequality, the sign of the statistic needs to be in the direction of the alternative
    print("We can reject the null hypothesis (p-value: ",p_value,")")
else:
    print("We cannot reject the null hypothesis (p-value: ",p_value,")")

print("Mean revenue for all movies:",movies["revenue"].mean())
print("Mean revenue for sample 1:",s1["revenue"].mean())
print("Mean revenue sample 2:",s2["revenue"].mean())
```

```python
"""
Is there a correlation between user ratings and critic ratings?

Pearson Correlation between user ratings and critic ratings for each movie
"""
```

```python
print("Person correlation: ", st.pearsonr(movies["vote_average"],movies["metacritic_score"])[0])
```

```python
x = movies[["vote_average"]]
y = movies["metacritic_score"]

model = linear_model.LinearRegression()
result = model.fit(x,y)
regression_line = result.intercept_ + result.coef_[0]*x

plt.scatter(movies["vote_average"],movies["metacritic_score"])
plt.plot(x,regression_line, c= "orange")
plt.xlabel("User ratings")
plt.ylabel("Metacritic ratings")
plt.show()

model.score(x,y)
```
